```python
import tensorflow as tf

model = tf.keras.models.load_model("/content/drive/MyDrive/EfficientNetB4_6000_tuning10.h5")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def img_predict(fname):
    img = tf.keras.utils.load_img(fname, target_size=(224, 224))
    plt.imshow(img)
    
    x=tf.keras.utils.img_to_array(img)
    images = np.expand_dims(x, axis=0)
    
    pred = model.predict(images, batch_size=10)
    predict = np.argmax(pred, axis=1)
    target = ['증상없음', '초기', '미성숙', '성숙']   # EfficientNet
    print(target[predict[0]])

    for i in range(len(pred[0])):
      print(f'{target[i]}일 가능성은 {round(pred[0][i] * 100)}% 입니다.')
```

```python
img_predict('/content/drive/MyDrive/백내장 이미지 샘플/train/미성숙/미성숙1.jpg')
```
